```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import time
import pandas as pd
import tqdm
import pickle

```

```python

# please enter your own hf token
hf_token = ""
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf", token= hf_token )
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf", token= hf_token, torch_dtype=torch.bfloat16)#, attn_implementation="flash_attention_2")
```

```python
def create_system_prompt(belief_system, topic, debater, opposition):
    system_prompt = f"""
    The task and setting:
    In the following, you will participate in a debate about a given topic. The debate will consist of 2 rounds, one starting round and two responsive rounds. In the starting round you should provide 3 points reflecting your beliefs. In the two following responsive rounds you will be provided with counter arguments to your point of belief that you need to analyze and address accordingly. The opposing side will do the same. You will be presented with a topic of conversation, a description reflecting your belief system, and rules regarding the debate with an opposing side. Your goal is to persuasively present your viewpoints while addressing the points raised by the opposing side.

    The topic of conversation:
    \"\"\"{topic}\"\"\"

    You name is:
    {debater}

    You oppositions name is:
    {opposition}


    Your belief system:
    \"\"\"{belief_system}\"\"\"



    How to debate:
    Rule 1. List the arguments of the opposition in their last response.  If there was so far no response from {opposition} ignore this rule.
    Rule 2. Bring forward points reflecting your beliefs.
    Rule 3. Counter the arguments of {opposition}. If there was so far no response from {opposition} ignore this rule.
    Rule 4. Speak directly to {opposition}
    Rule 5. Engage in a back-and-forth exchange of arguments and counterarguments for each round. You should aim to persuasively present your viewpoints while addressing the points raised by {opposition}. Remember, it’s not necessary to agree!
    Rule 6. It’s important that your arguments AND counterarguments reflect the belief system provided.


    Example format:
    \"\"\"
    Oppositions arguments:
    1. Birth control kills a potential live.
    2. Free birth control leads to casualty of the topic.
    3. Every live is sacred and beautiful.

    While you are correct that a potential life is being "killed",
    I would argue that every woman's right to decide what to do with her body outweighs the negative. That leads me to yours
    Argument that every life is sacred and beautiful. We need to see women becoming unintentionally and perhaps even porous
    The sexual act may not have been consensual. Let's imagine the worst case scenario, a financial weak woman is being raped and doesnt wont the child of
    her abuser. In my opinion, the damage could be minimized by at least helping this woman with the burden of raising a child she never wanted.
    I understand where you're coming from when you argue that the topic is becoming too casual, but the impact on a woman's body is huge
    and I don't think a woman would do this lightly several times. So I strongly disagree. Another important point for abortion in my opinion is
    that criminalizing abortion does not stop abortions, it just makes them less safe.
    \"\"\"


    """

    return system_prompt



```

```python
def debate_round(history, model, tokenizer):
    now = time.time()
    encodeds = tokenizer.apply_chat_template(history, return_tensors="pt")
    encodeds = encodeds.to("cuda:0")
    answer = model.generate(encodeds, max_new_tokens = 1000)
    answer_extracted = tokenizer.decode(answer[0], skip_special_tokens=True )
    new_answer = new_answer = answer_extracted.split("[/INST]")[-1]
    history.append({"role": "assistant", "content": new_answer})
    time_llama.append(time.time()-now)
    torch.cuda.empty_cache()
    del encodeds
    return history
```

```python
time_llama = []
history_df = pd.DataFrame(columns=["run1", "run2", "run3"])
rounds = 2
runs = 4
for run in tqdm.tqdm(range(1,runs)):
  for index, row in base_prompts.iterrows():
      debaters = row["Names"].split(" / ")
      belief_systems = [row["Stance1"], row["Stance2"]]
      topic = row.Topic
      system_prompts = {}
      for i in range(len(belief_systems)):
          bs = belief_systems[i]
          opposition = debaters.copy()
          opposition.pop(i)
          system_prompts[debaters[i]] = create_system_prompt(bs, topic, debaters[i], opposition[0])
      prompt = f"What do you think about {topic}?"
      history = [{"role": "system", "content": system_prompts[debaters[0]]},
                {"role": "user", "content": prompt}]
      iteration = 0
      for round in tqdm.tqdm(range(rounds)):
          for debater in debaters:
              # adjust the system prompt based on whos turn it is
              history[0]["content"] = system_prompts[debater]
              if iteration != 0:
                  history.append({"role":"user", "content":f'Now its your turn {debater}'})
              # get the new history (all the debate rounds)
              history = debate_round(history, model, tokenizer)
              iteration += 1
      tmp_run = f'run{run}'
      history_df.loc[index, tmp_run] = history
```

```python
history_df.to_csv("/content/histories_llama_few.csv")
```
